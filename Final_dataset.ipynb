## df3 Transform to df4

```python
df4 = pd.DataFrame(columns=['Zip', 'Good', 'Fair', 'Poor'])

for index, row in df3.iterrows():
    good = 0
    fair = 0
    poor = 0
    
    old_good = str(row['Good'])
    old_fair = str(row['Fair'])
    old_poor = str(row['Poor'])

    if "Good" in old_good:
        val = old_good.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        good = int(val)
    elif "Good" in old_fair:
        val = old_fair.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        good = int(val)
    elif "Good" in old_poor:
        val = old_poor.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        good = int(val)
    else:
        good = 0

    if "Fair" in old_good:
        val = old_good.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        fair = int(val)
    elif "Fair" in old_fair:
        val = old_fair.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        fair = int(val)
    elif "Fair" in old_poor:
        val = old_poor.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        fair = int(val)
    else:
        fair = 0

    if "Poor" in old_good:
        val = old_good.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        poor = int(val)
    elif "Poor" in old_fair:
        val = old_fair.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        poor = int(val)
    elif "Poor" in old_poor:
        val = old_poor.split(":")[1]
        if "{" in val:
            val = val.replace("{", "")
        elif "}" in val:
            val = val.replace("}", "")
        poor = int(val)
    else:
        poor = 0

    df4 = df4.append({'Zip': row['zip'], 'Good': good, 'Fair': fair, 'Poor': poor}, ignore_index=True)

df4
```
